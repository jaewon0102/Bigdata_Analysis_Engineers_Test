#### 데이터 세트의 TAX 칼럼이 TAX 칼럼의 중위값보다 큰 데이터를 대상으로, CHAS 칼럼과 RAD칼럼 순으로 그룹을 지은 후 각 그룹의 데이터 개수를 구하시오. 단, CHAS, RAD 칼럼별 데이터 개수는 COUNT라는 칼럼으로 출력한다.

[ boston.csv ] 활용

```python
import pandas as pd
```

```python
df = pd.read_csv('./boston.csv')
print(df.shape) # 데이터 프레임 행과 열 확인
df.head()
```

```python
df_m=df['TAX'].median() # df_m을 TAX의 중앙값변수로 지정하기
df_m
```

```python
df[df['TAX']>df_m] # 중앙값 보다 큰 값 여부 확인
```

```python
# TAX가 중앙값보다 큰 데이터 프레임들 중 CHAS와 RAD만 뽑아오기
df_new=df[df['TAX']>df_m][['CHAS','RAD']]
df_new
```

```python
#그룹화 -> 칼럼 두개의 결과에 count만 하면되기때문에 둘 중 하나만 선택하면 됨
df_new2=df_new.groupby(['CHAS','RAD'])[['RAD']].count()
df_new2.columns=['COUNT'] # 칼럼명 지정하기
df_new2
```
