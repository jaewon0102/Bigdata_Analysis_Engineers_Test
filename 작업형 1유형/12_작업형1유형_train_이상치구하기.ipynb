이상치를 찾아라
<br>데이터에서 IQR을 활용해 Fare컬럼의 이상치를 찾고, 이상치 데이터의 여성 수를 구하시오

[ train.csv ] -> 활용

```python
import pandas as pd
```

```python
df = pd.read_csv('./train.csv')
print(df.shape) # dataset 행과 열 확인
df.head()
```

```python
Q1 = df['Fare'].quantile(.25)
Q3 = df['Fare'].quantile(.75)
IQR = Q3 - Q1

Q1 , Q3, IQR
```

```python
data1 = Q1 - (IQR*1.5)
data2 = Q3 + (IQR*1.5)
data1, data2
```

```python
df1=df[(df['Fare']<data1)|(df['Fare']>data2)] # 이상치 범위 내 값들 추출하기
df1
```

```python
result = df1[df1['Sex']=='female']['Sex'].count()
result
```
