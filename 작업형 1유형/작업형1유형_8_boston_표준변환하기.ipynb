#### 데이터 세트의 DIS 칼럼을 표준화 척도(Standard Scale)로 변환한 후, 0.4보다 크면서 0.6보다 작은 값들에 대한 평균을 구하시오. 단, 소수점 셋째 자리에서 반올림하여 소수점 둘째 자리까지 출력하시오.

[ boston.csv ] -> 활용

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler # 표준화 척도 standardscaler 변환
```

```python
df = pd.read_csv('./boston.csv')
print(df.shape) # 데이터 셋 행과 열 확인
df.head()
```

```python
scaler = StandardScaler()

# 표준화척도로 변환해서 데이터프레임으로 만들고, 원하는 dis 컬럼만 뽑아오기
df_s = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)[['DIS']]
df_s
```

```python
df_s1=df_s[(df_s['DIS']>0.4)&(df_s['DIS']<0.6)]
df_s1
```

```python
result = round(df_s1['DIS'].mean(),2) # 2는 반올림해서 두번째자리까지 출력하는것
result
```
