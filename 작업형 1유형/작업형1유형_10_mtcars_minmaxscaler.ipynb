#### 자동차 데이터 셋에서 qsec 칼럼을 Min-Max Scale 로 변환 후 0.5보다 큰 값을 가지는 레코드(row) 수를 묻는 문제

[ mtcars.csv ] -> 활용

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('./mtcars.csv')
print(df.shape) # dataset 행과 열 확인
df.head()
```

```python
scaler = MinMaxScaler()

a = df[['qsec']] # df에서 문자로된 것도 있어서 필요한 컬럼을 따로 설정하여 변환하기
df_m = pd.DataFrame(scaler.fit_transform(a),columns=['qsec'])
df_m
```

```python
result = df_m[df_m['qsec'] > 0.5]['qsec'].count()
result
```
