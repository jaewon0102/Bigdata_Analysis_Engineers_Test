#### 데이터 세트의 전체 칼럼에 대해서 중복을 제거한 유니크한 값을 구한 후, 칼럼별로 유니크한 값의 개수를 기준으로 평균값을 구하시오

[ boston.csv ]-> 활용

```python
import pandas as pd
```

```python
df = pd.read_csv('./boston.csv')
print(df.shape) # dataset 행과 열 확인하기
df.head()
```

```python
df_c = df.columns
df_c
```

```python
df['CRIM'].unique()
```

```python
pd.DataFrame(df['CRIM'].unique()).count()
```

```python
# 뒤에 다른 변수들도 수가 딱 떨어질 수 있게
int(pd.DataFrame(df['CRIM'].unique()).count())
```

##### 첫번째 방법

```python
# 컬럼 하나하나 다 치기
print((int(pd.DataFrame(df['CRIM'].unique()).count()) +
    int(pd.DataFrame(df['ZN'].unique()).count()) +
    int(pd.DataFrame(df['INDUS'].unique()).count()) +
     int(pd.DataFrame(df['CHAS'].unique()).count()) +
      int(pd.DataFrame(df['NOX'].unique()).count()) +
      int(pd.DataFrame(df['RM'].unique()).count()) +
      int(pd.DataFrame(df['AGE'].unique()).count()) +
      int(pd.DataFrame(df['DIS'].unique()).count()) +
      int(pd.DataFrame(df['RAD'].unique()).count()) +
      int(pd.DataFrame(df['TAX'].unique()).count()) +
      int(pd.DataFrame(df['PTRATIO'].unique()).count()) +
      int(pd.DataFrame(df['B'].unique()).count()) +
      int(pd.DataFrame(df['LSTAT'].unique()).count()) +
      int(pd.DataFrame(df['MEDV'].unique()).count())
      ) / df_c.size) # df_c변수의 개수를 세어줌
```

##### 두번째 방법

```python
# 최초 합계는 0으로 설정하기
sum = 0

# df 변수의 컬럼별로 유니크한 값 개수를 구한 후, sum 변슈에 더하는 for문
for 한개씩 in df_c:
    sum = sum+int(pd.DataFrame(df[한개씩].unique()).count())

sum
```

```python
result = sum / df_c.size
result
```
