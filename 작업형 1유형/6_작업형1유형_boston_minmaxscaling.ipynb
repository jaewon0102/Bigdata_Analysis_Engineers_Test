#### 데이터 세트의 MEDV 칼럼을 최소최대 척도(Min-Max Scale)로 변환한 후 0.5보다 큰 값을 가지는 레코드 수를 구하시오.

[ boston.csv ] -> 활용

```python
import pandas as pd
```

```python
df = pd.read_csv('./boston.csv')
print(df.shape) # 데이터프레임 행과 열 확인
df.head()
```

```python
# 머신러닝 라이브러리인 sklearn에서 MinMaxScaler import 하기
from sklearn.preprocessing import MinMaxScaler
```

#### 첫번째 방법

```python
scaler = MinMaxScaler()
#최소최대 척도로 변환하기
df_minmax=pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df_minmax
```

```python
# 0.5보다 큰 값을 가지는 컬럼 확인하기
df_minmax[df_minmax['MEDV']>0.5]['MEDV']
```

```python
result=df_minmax[df_minmax['MEDV']>0.5]['MEDV'].count()
result
```
