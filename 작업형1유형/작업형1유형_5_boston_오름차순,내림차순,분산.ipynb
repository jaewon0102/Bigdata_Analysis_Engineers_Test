#### 데이터 세트의 TAX 칼럼을 오름차순으로 정렬한 결과와 내림차순으로 정렬한 결과를 각각 구한다. 그리고 각 순번에 맞는 오름차순 값과 내림차순 값의 차이를 구하여 분산 값을 출력한다.

[ boston.csv ] -> 활용

```python
import pandas as pd
```

```python
df = pd.read_csv('./boston.csv')
print(df.shape) # 데이터프레임 행과 열 확인하기
df.head()
```

```python
# 오름차순 (작은 값 부터)
df_t=df.sort_values(by='TAX',ascending = True)[['TAX']]
df_t
```

```python
# 내림차순 (큰 값 부터)
df_d = df.sort_values(by='TAX',ascending = False)[['TAX']]
df_d
```

```python
# index가 섞여서 저장된걸 reset해서 차례대로 정리하기
# 이 과정을 거치지않으면 두 값의 차이를 구할때 제대로 된 값이 나오지 않음
df_t.reset_index(drop = True, inplace = True)
df_d.reset_index(drop = True, inplace = True)
```

```python
# concat -> axis=1 은 행 단위가 아닌 컬럼 단위로 붙이기 때문
df_concat = pd.concat([df_t,df_d],axis=1)
df_concat
```

##### 첫번째 방법

```python
df_concat['diff']=abs(df_concat.iloc[:,0]-df_concat.iloc[:,1])
df_concat
```

```python
df_concat['diff'].var()
```

##### 두번째 방법

```python
print(abs(df_t-df_d).var())
```

##### 첫번째, 두번째 방법 모두 동일한 값이 나오지만, 실제 시험에선 '숫자만' 나오도록 하기 때문에 두번째 방법을 사용할시,
##### </br>df.sort_values(by='TAX',ascending = False)['TAX']  -> 뒤에 'TAX'의 대괄호를 한번만 써서 결괏값을 도출할 것
