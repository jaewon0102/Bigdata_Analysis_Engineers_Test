주어진 Dataset에서 CRIM값이 가장 큰 10개의 지역을 구하고 10개의 지역의 CRIM값을 그 중 가장 작은 값으로 대체한다. AGE 칼럼 값이 80이상인 대체된 CRIM 평균 값을 구하라. (boston)

```python
import pandas as pd
```

```python
df=pd.read_csv('./bigData-main/boston.csv')
df
```

```python
# 'CRIM' 을 기준으로 ascending이 False면 큰 숫자부터 내려감 (True면 작은 숫자부터 올라감)
df_c=df.sort_values(by='CRIM',ascending = False)
```

#### 1번째 방법

```python
# 10번째로 큰 값이 뭔지 확인
df_c.head(10)
```

```python
# 9번째까지 10번째로 큰 값으로 바꾸기
df_c['CRIM'].iloc[:9]=25.9406
```

```python
# 상위 10개만 바뀐게 맞는지 확인
df_c.head(11)
```

#### 2번째 방법

```python
min_crim = df_c[:10]['CRIM'].min()
```

```python
df_c.iloc[:10, 0] = min_crim
df_c.head(11)
```

```python
# 생략가능
# 필터링 한 것들 중 age가 80이상인 것을 추출 후 확인
df_a=df_c[df_c['AGE']>=80][['AGE']]
df_a
```

```python
# 필터링 한 것들 중 age가 80이상인 CRIM을 추출
df_c[df_c['AGE']>=80]['CRIM']
```

```python
# 결과
result = df_c.loc[df_c['AGE']>=80, 'CRIM'].mean()
result
```
